```python
 highest_salary = data['salary_usd'].max() 
print (f'The highest salary in USD is {highest_salary}')
print(data[data['salary_usd'] == data['salary_usd']].max())
#Robotics engineer has the highest paying job in the transportantion industry.
```

```python
#Grouping the jobs title and the salary will make us see all the trends.
salary_to_job_title = data.groupby('job_title')['salary_usd'].sum() 
print(salary_to_job_title)
```

```python
plt.figure(figsize=(12,6))
salary_to_job_title.plot(kind="bar",color="blue")
plt.title('Total Jobs with thier Salary in USD', fontsize=14)
plt.xlabel('Job Title', fontsize=12)
plt.ylabel('Salary in USD', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.05)
plt.tight_layout()
plt.show()
```

```python
#sum of each job on the raise
sum_of_job_title = data['job_title'].value_counts()
print(f'The sum of each job is {sum_of_job_title}')
#The most popular job is Machine Learning with makes sense since the raise of BIG DATA it raises the need to have algorithms to manage the data. 
```

```python
plt.figure(figsize =(12, 4))
sum_of_job_title.plot(kind = 'bar', color = 'red')
plt.title("JOBS OFFERED BY AI.", fontsize = 14)
plt.xlabel("Job title", fontsize = 14)
plt.ylabel("Number of jobs offered", fontsize = 14)
plt.grid(axis = "y" , alpha = 0.05)
plt.tight_layout
plt.show()
```
